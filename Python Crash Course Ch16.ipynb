```python
import csv
```

```python
filename = 'sitka_weather_07-2014.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)
    first_row = next(reader)
    
    for index, column_header in enumerate(header_row):
        print(index, column_header)
    
    print(first_row)
    
    highs = []

    for row in reader:
        high = int(row[1])
        highs.append(high)
        
    print(highs)
```

```python
from matplotlib import pyplot as plt

fig = plt.figure(dpi=128, figsize=(5, 3))
plt.plot(highs, c= 'pink')

plt.title("Dailiy High Temp", fontsize=24)
plt.xlabel("dates", fontsize = 10)
plt.ylabel("Temp", fontsize=15)
plt.tick_params(axis='both', which='major', labelsize=16)

plt.show()
```

```python
f_name = 'sitka_weather_07-2014.csv'
with open(f_name) as a_file:
    a_reader = csv.reader(a_file)
    header_row = next(a_reader)
    
    h_temps, dates = [], []
    for r in a_reader:
        h_temp = int(r[1])
        h_temps.append(h_temp)
        
        date = dt.strptime(r[0], '%Y-%m-%d')
        dates.append(date)
        
    
a_figure = plt.figure(dpi=128, figsize=(5,3))
plt.plot(dates, h_temps, c='pink')
plt.title("Daily High Temp (F)", fontsize=12)
plt.ylabel("Temp", fontsize=8)
a_figure.autofmt_xdate()
plt.tick_params(axis='both', which='major', labelsize =7)
plt.show()
```

```python
with open('death_valley_2014.csv') as sitka_file:
    reader_sitka = csv.reader(sitka_file)
    header_row = next(reader_sitka)
#     print(header_row)
    
    high_temps, low_temps, mean_temps, dates = [], [], [], []
    for row in reader_sitka:
        try:
            high_t = int(row[1])
            low_t = int(row[3])
            mean_t = int(row[2])        
            date = dt.strptime(row[0], '%Y-%m-%d')
        except ValueError:
            continue
        else:
            high_temps.append(high_t)
            low_temps.append(low_t)
            mean_temps.append(mean_t)
            dates.append(date)
        
sitka_temp_fig = plt.figure(dpi = 128, figsize=(10,3))     
plt.plot(dates, high_temps, c = 'c')
plt.plot(dates, low_temps, c= 'r')

plt.fill_between(dates, high_temps, mean_temps, facecolor='c', alpha = 0.5)
plt.fill_between(dates, low_temps, mean_temps, facecolor = 'r', alpha = 0.4)

plt.title("Max and Min Temp \n Death Valley")
plt.xlabel("date")
plt.ylabel("Temp")


plt.show()
```
